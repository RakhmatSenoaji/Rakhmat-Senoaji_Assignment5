 2️⃣
• Konversi data ke GeoDataFrame dan pastikan CRS sudah benar.
Deskripsi:
1. Konversi GeoJSON ke GeoDataFrame.
2. Set CRS ke EPSG:4326 jika belum ada.

```python
#Konversi data geoJSON ke GeoDataFrame dan set CRS ke EPSG:4326
gdf = gpd.GeoDataFrame.from_features(data)

gdf = gdf.set_crs(epsg=4326)

gdf.head()
```

3️⃣
• Buat zona buffer (penyangga) sejauh 100 meter di sekitar setiap polygon pada data 
spasial. Buffer sering digunakan untuk analisis zona pengaruh atau area aman di sekitar 
objek spasial.
Soal:
1. Reproject data ke CRS meter (EPSG:3857).
2. Buat buffer 100 meter untuk setiap polygon.
3. Kembalikan ke EPSG:4326 dan export hasilnya ke GeoJSON.

```python
gdf_3857 = gdf.to_crs(epsg=3857)

buffer = gdf_3857.buffer(distance=100)

buffer.head()

m = folium.Map(location=[-6.95, 107.68], zoom_start=12)
gj = folium.GeoJson(buffer.to_crs(epsg=4326), name="geojson").add_to(m)
m.fit_bounds(gj.get_bounds())
m
```
